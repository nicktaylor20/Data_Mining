```python
import numpy as np
import pandas as pd
```

```python
#Creating array with random whole numbers 0-500
array = np.random.randint(0,501,500)
```

```python
print(array)
```

```python
#Reshaping array to 10,50
a = array.reshape(10,50)
```

```python
a.shape
```

```python
print(a)
```

```python
#loading array to a dataframe 
df = pd.DataFrame(a)
```

```python
print(df)
```

```python
#Sum of each row
df.sum(1)
```

```python
#sum of each column
df.sum(0)
```

```python
df.cumsum()
```

```python
#Min values
df.min()
```

```python
#Max Values
df.max(0)
```

```python
#Average values
df.mean(0)
```

```python
#Standard deviation
df.std()
```
