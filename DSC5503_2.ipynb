# Rank number: 3 Multi-Plot

```python
#Multiplot 
x1 = df['deaths']
x2 = df['deaths']

y1 = df['date']
y2 = df['state']

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Covid Death Toll by Date and State')
figsize=(30,30)

ax1.plot(x1, y1)
ax1.set_ylabel('Dates')

ax2.barh(y2, x2)
ax2.set_xlabel('Number of Deaths')
ax2.set_ylabel('State')

plt.show()


```

# Rank Number: 2 Bar Chart

```python
plt.barh(df['state'], df['deaths'])
plt.title('US Covid Deaths in Southern States')
plt.xlabel('Number of Deaths')
plt.ylabel('State')
```
